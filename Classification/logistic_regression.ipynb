```python
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
```

## ロジスティック回帰

```python
df = sns.load_dataset('titanic')
df.describe()
```

```python
df = df.dropna()
df
```

```python
X = df.loc[:, (df.columns != 'survived') & (df.columns != 'alive')]
X = pd.get_dummies(X, drop_first=True)
y = df['survived']

# hold-out
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)
log_loss(y_test, y_pred_proba)
```

```python
# 最急降下法で計算された最適なθの値が入っている。
model.coef_
```

```python
# ラベルの確率p(X)を取得
y_pred_proba
```

## 多クラス分類のロジスティック回帰

```python
df = sns.load_dataset('iris')
df.head()
```

### OVR

```python
y_col = 'species'
y = df[y_col]
X = df.loc[:, df.columns != y_col]

# hold-out
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

model = LogisticRegression(penalty='none', multi_class='ovr')
model.fit(X_train, y_train)
y_pred_ovr = model.predict_proba(X_test)
print(model.classes_)
print(model.intercept_)
print(model.coef_)
```

### multinomial

```python
model = LogisticRegression(penalty='none', multi_class='multinomial')
model.fit(X_train, y_train)
y_pred_mn = model.predict_proba(X_test)
print(model.classes_)
print(model.intercept_)
print(model.coef_)
```
